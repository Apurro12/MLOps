```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plotear_clasificador(clf):
    
    title = color == clf.predict(X)
        
        
    plt.scatter(X[:,0], X[:,1], c = clf.predict(X))
    plt.title(title.mean())
    plt.show()
```

```python
X = 10*np.random.random(size = (1000,2))
```

```python
color = (X[:,0] - 5 )**2 + (X[:,1] - 5)**2 < 8
```

```python
plt.scatter(X[:,0], X[:,1], c = color)
plt.show()
```

```python
class model:
    
    def __init__(self, x_min = -np.inf , x_max = np.inf, y_min = -np.inf, y_max = np.inf):
        
        self.x_min = x_min
        self.x_max = x_max
        self.y_min = y_min
        self.y_max = y_max
        
        
    def predict(self, X):
        
        X_th = (X[:,0] >= self.x_min) & (X[:,0] <= self.x_max)
        Y_th = (X[:,1] >= self.y_min) & (X[:,1] <= self.y_max)
        
        return X_th & Y_th
```

```python
class ensamble:
    
    def __init__(self,list_models):
        
        self.list_models = list_models
        
    def predict(self, X):

        X_th_final = np.array([True]*len(X))
        y_th_final = np.array([True]*len(X))

        for model in self.list_models:

            X_th = (X[:,0] >= model.x_min) & (X[:,0] <= model.x_max)
            Y_th = (X[:,1] >= model.y_min) & (X[:,1] <= model.y_max)

            X_th_final = X_th_final & X_th
            y_th_final = y_th_final & Y_th




        return X_th_final & y_th_final
```

```python
model1 = model(x_min = 2)
```

```python
plotear_clasificador(model1)
```

```python
model2 = model(x_max = 8)
plotear_clasificador(model2)
```

```python
model3 = model(y_min = 2)
plotear_clasificador(model3)
```

```python
model4 = model(y_max = 8)
plotear_clasificador(model4)
```

```python
plt.scatter(X[:,0], X[:,1], c = color)
plt.show()
```

```python
#El ensamble es el mejor de todos los modelos
plotear_clasificador(ensamble([model1, model2, model3, model4]))
```
